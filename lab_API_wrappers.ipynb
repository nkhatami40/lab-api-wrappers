It can be seen that the name of the songs referred to as 'name' is in the nested dictionary with the key of track.

```python
#getting the names of the songs in the playlist
tracks_playlist=[]
for i in range(0,len(playlist['items'])):
    title = playlist['items'][i]['track']['name']
    tracks_playlist.append(title)
tracks_playlist
```

```python
#Getting the track ids
track_ids = []
for i in range(0,len(playlist['items'])):
    track_id = playlist['items'][i]['track']['id']
    track_ids.append(track_id)
track_ids
```
